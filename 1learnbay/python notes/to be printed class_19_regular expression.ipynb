```python
#keywords used for regular expresion

'''
.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)
'''
```

# initially we must import re 
import re

```python
text_to_search = '''
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ\s
321-555-4321
1234567890
Ha HaHa
MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
khanafsaan11.com
321-555-4321
123.555.1234
123*555*-1234
123.555.1234
800-555-1234
900-555-1234
Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
Mr_hello
'''
```

```python
# what you want to match
pattern = re.compile('abc')

# from which text you have to match
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# dot(.) (any char except new line)

# what you want to match
pattern = re.compile(r'.')### we use r bcz it is rawstring for better results

# from which text you have to match
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \d (0-9)[any number from 0-9]

# what you want to match
pattern = re.compile(r'\d')

# from which text you have to match
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \D not(0-9)everthing except 0-9
# what you want to match
pattern = re.compile(r'\D')

# from which text you have to match
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \w - (a-z . A-Z , 0-9 , _) 

# what you want to match
pattern = re.compile(r'\w')

# from which text you have to match
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \W - not a word character(special char and newline)

pattern = re.compile(r'\W')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \s - whitspace ( space , tab , \n)

pattern = re.compile(r'\s')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \S - not a whitespace

pattern = re.compile(r'\S')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# example number 1

# extract all the 3 digit numbers

text_to_search = '''
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ\s
321-555-4321
1234567890
Ha HaHa
MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
khanafsaan11.com
321-555-4321
123.555.1234
123*555*-1234
123.555.1234
800-555-1234
900-555-1234
Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
Mr_hello
'''
```

```python
# Extract all the phone numbers from a text 3.3.4

pattern = re.compile(r'\d\d\d\.\d\d\d\.\d\d\d\d')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# [] - matches char in bracktes

pattern = re.compile(r'[ab]')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
text_to_search = '''
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ\s
321-555-4321
1234567890
Ha HaHa
MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
khanafsaan11.com
321-555-4321
123.555.1234
123*555*-1234
123.555.1234
800-555-1234
900-555-1234
Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
Mr_hello
'''
```

```python
# step 5 - 

pattern = re.compile(r'Mr\.\s[A-Z]')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```
